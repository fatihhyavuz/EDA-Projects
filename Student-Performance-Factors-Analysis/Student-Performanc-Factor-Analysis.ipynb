### Column Descriptions

1. **Hours_Studied**: Number of hours the student studied.
2. **Attendance**: Attendance rate or percentage of classes attended by the student.
3. **Parental_Involvement**: Level of parental involvement in the student's academic life.
4. **Access_to_Resources**: Availability of educational resources for the student.
5. **Extracurricular_Activities**: Participation in extracurricular activities.
6. **Sleep_Hours**: Average number of hours the student sleeps per night.
7. **Previous_Scores**: Previous academic scores or grades of the student.
8. **Motivation_Level**: The student’s level of motivation toward academic goals.
9. **Internet_Access**: Whether the student has access to the internet (Yes/No).
10. **Tutoring_Sessions**: Number of tutoring or mentoring sessions attended by the student.
11. **Family_Income**: Income level of the student's family.
12. **Teacher_Quality**: Perceived quality of teachers (e.g., rated by experience or feedback).
13. **School_Type**: Type of school the student attends (e.g., public, private).
14. **Peer_Influence**: Level of influence from peers on the student's academic performance.
15. **Physical_Activity**: Amount of physical activity the student participates in.
16. **Learning_Disabilities**: Whether the student has any learning disabilities (Yes/No).
17. **Parental_Education_Level**: Education level of the student’s parents.
18. **Distance_from_Home**: Distance of the school from the student's home.
19. **Gender**: Gender of the student.
20. **Exam_Score**: The student's score on a recent exam.

## <font color='midnightblue'> <b>Determining the Categoric Columns </b><font color='black'> 

```python
# Selected columns with dtype values of category, bool, and object
cat_cols = [col for col in df.columns if str(df[col].dtypes) in ["object", "category", "bool"]]

# However, some columns with dtype int or float might also be categorical, e.g., survived
num_but_cat = [col for col in df.columns if df[col].nunique() < 6 and df[col].dtypes in ["int64", "float"]]

# There may be variables with dtype category, bool, and object that are not actually categorical, e.g., name, e-mail
cat_but_car = [col for col in df.columns if df[col].nunique() > 6 and str(df[col].dtypes) in ["object", "category", "bool"]]

# Exclude columns containing "percentage", "rate", or "ratio"
cat_cols = [col for col in cat_cols if "percentage" not in col and "rate" not in col and "ratio" not in col]

# Combine cat_cols and num_but_cat
cat_cols = cat_cols + num_but_cat

# If data comes from cat_but_car, it needs to be excluded
cat_cols = [col for col in cat_cols if col not in cat_but_car]

# Print the final list of categorical columns
cat_cols
```

<a id='understanding-dataset'></a>
## <font color='midnightblue'> <b>Determining  the Numeric Colums </b><font color='black'> 

```python
num_cols = [i for i in df.columns if i  not in cat_cols]
```

## <font color='midnightblue'> <b>Function to Get Summary Information For Specified Categorical Columns </b><font color='black'> 

```python
def cat_summary(dataframe, col_name):
    print(pd.DataFrame({
        col_name: dataframe[col_name].value_counts(),  # Count of unique values
        "Ratio": 100 * dataframe[col_name].value_counts() / len(dataframe)  # Percentage of each unique value
    }))
    print("------------------------------------------------------------------")

for col in cat_cols:
    cat_summary(df, col)
```
